*3.3* (10 pts) Compute the incidence proportion (number of deaths each day per 100,000 population) for each county, you may save this as a new dataset.  You can calculate the incidence for a given county by taking the difference of the cumulative cases for a day with the previous day.  Notice that this dataset is quite large, one trick to make this faster is to use the fips code and date as indexing variables and use `sort_index` to sort the indices before applying groupby operations.  For full points, you should use `groupby` and `diff` for this operation.

```python
new_Deaths['Date'] = pd.to_datetime(new_Deaths['Date'])
```

```python
new_Deaths['Date'] = new_Deaths['Date'].dt.strftime('%m/%d/%Y')
```

```python
FD_indexdf = new_Deaths.set_index(['fips','Date'])
```

```python
data = FD_indexdf.groupby(['county_name', 'Date']).sum()
```

```python
data['diff'] = data['deaths'].diff()
```

```python
data = data[data['diff'] >= 0]
```

```python
data['deathPerday'] = data['diff']/data['Population'] * 100000
```

*3.4* (15 pts) Compute the incidence proportion per week (per 100,000 population), which is the sum of the incidences for all days in a week.  Hint: use Periods in Pandas.  A week is defined to be Sunday through Saturday.  Which is the week with the highest mean incidence proportions (where the mean is taken over all the counties)?  You can either sum the incidence counts within the week, or take a difference of the cumulative cases between consecutive Sundays, and these should produce the same thing.

```python
new_Deaths['Date2'] = pd.PeriodIndex(new_Deaths['Date'], freq='W-SAT') 
```

```python
Data2 = new_Deaths.set_index(['fips', 'Date2'])
```

```python
data_deaths = Data2.groupby(['county_name', 'Date2']).sum()
```

```python
data_deaths['diff'] = data_deaths['deaths'].diff()
```

```python
data_deaths.head(10)
```

```python
data_deaths['deathPerweek'] = data_deaths['diff']/data_deaths['Population'] * 100000
```

```python
data_deaths = data_deaths[data_deaths['diff'] >= 0]
```

```python
# week with the highest mean incidence proportions(per week)

data_deaths.groupby('Date2').sum()[data_deaths.groupby('Date2').sum()['deathPerweek'] == data_deaths.groupby('Date2').sum()['deathPerweek'].max()]
```
